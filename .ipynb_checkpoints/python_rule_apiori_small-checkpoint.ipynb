```python
# !pip install mlxtend

import numpy as np
import pandas as pd
import seaborn as sns


from mlxtend.frequent_patterns import apriori,fpgrowth
from mlxtend.frequent_patterns import association_rules

from mlxtend.preprocessing import TransactionEncoder
```

```python
basketData = pd.read_csv('df_small_id.csv')
```

```python
trans = basketData.groupby(['t_id'])['item'].apply(list).values.tolist()

trans
```

```python
te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.0145, use_colnames=True) 
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.sort_values(by=['support','confidence'], ascending=False)[0:9]
```

```python
frequent_itemsets_fp = fpgrowth(df, min_support=0.0145, use_colnames=True) 
```
