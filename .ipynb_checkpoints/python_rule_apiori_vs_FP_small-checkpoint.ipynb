```python
# !pip install mlxtend

import numpy as np
import pandas as pd
import seaborn as sns


from mlxtend.frequent_patterns import fpgrowth,apriori
from mlxtend.frequent_patterns import association_rules

from mlxtend.preprocessing import TransactionEncoder
```

```python
basketData = pd.read_csv('df_small_id.csv')
```

```python
trans = basketData.groupby(['t_id'])['item'].apply(list).values.tolist()

trans
```

```python
te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
FP_frequent_itemsets = fpgrowth(df, min_support=0.0145)
```

```python
rules = association_rules(FP_frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.sort_values(by=['support','confidence'], ascending=False)[0:9]
```

```python
apriori_frequent_itemsets = apriori(df, min_support=0.0145)
```

```python
rules_2 = association_rules(apriori_frequent_itemsets, metric="lift", min_threshold=1)
rules_2
```

**Apriori:** Execute 100 loops, 10 times

```python
%timeit -n 100 -r 10 apriori(apriori_frequent_itemsets, min_support=0.0145)
```

**Apriori:** Execute 100 loops, 10 times with *low memory usage*

```python
%timeit -n 100 -r 10 apriori(apriori_frequent_itemsets, min_support=0.0145, low_memory=True)
```

**FP-Growth:** Execute 100 loops, 10 times

```python
%timeit -n 100 -r 10 fpgrowth(df, min_support=0.6)
```
