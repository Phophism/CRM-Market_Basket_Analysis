```python
import pandas as pd
```

```python
data = pd.read_csv('https://query.data.world/s/ydwrbk7oo5gw5f7tsljevsc4jsgcyk')
```

```python
data.head(5)
```

```python
data[['Dept', 'item']] = data['Dept'].str.split(':', 1, expand=True)
data
```

```python
data['t_id']=data.groupby(['POS Txn']).ngroup()
```

```python
df = data.drop(['Dept','ID','Sales U','POS Txn'],1)
df
df.to_csv('df_small.csv',float_format='%.15f',index=False)
```

```python
transaction = df.groupby(['t_id'])['item'].apply(list).values.tolist()
 
transaction 
```

```python
import csv 
with open("small.csv", 'w', newline='') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerows(transaction)
```
